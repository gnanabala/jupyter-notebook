###  Analytic workbench is the web based tool for performing different kinds of analyses. 

### This user interface is backed by a multi-agent system and each of the modules in the visual language corresponds to one agent in the backend


###  Analysis workflows are composed using a visual language based on a pipes-andfilters metaphor. <br>

### In this language the modules (filters) of the language represent Individual analysis steps and the connections (pipes) between the modules represent the data flow <br>

### This allows analysts even with no or little programming to design and execute complex data analysis tasks





** The workbench framework uses a SQLSpaces server as communication platform**<br>

**  The communication between the components of the system uses a simple tuplebased protocol, of which the two main elements are command tuples controlling the execution of a workflow and data tuples used for transferring data between the agents**
